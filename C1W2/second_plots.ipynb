```python
import scipy.linalg as la
import numpy as np
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])

answer = la.solve(a, b)
print(answer)
```
