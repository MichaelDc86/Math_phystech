data from wiki

$E[x]=0.5$

$D[x]=0.125$

generate sample with size=1000

```python
volume = 1000
sample = sts.arcsine.rvs(size=volume)
# len(sample)
# sample
```

histogram

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(0,1,1000)
y = 1 / math.pi * np.sqrt(x * (1 - x))

plt.plot(x, y, label='theoretical plot')

plt.plot(x, sts.arcsine.pdf(x), label='theoretical pdf')

plt.hist(sample, bins=30, density=True, range=(0, 1), label='sample')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
ax.legend(loc='best', frameon=False)
plt.show()
```

Generate 1000 samples with different volumes    

### n = 5

```python
n = 5
selected_avarage_arr_5 = np.array([])  #array for selected avarage of i-th sample
for i in range(1, 1001):
    sample_ = sts.arcsine.rvs(size=n) 
    selected_avarage = sum(sample_) / len(sample_)  # summ all elements and devides on their amount
    selected_avarage_arr_5 = np.append(selected_avarage_arr_5, selected_avarage)
# len(selected_avarage_arr_5)
# selected_avarage_arr
```

normal distributions parameters
$N(mu, sigma)$

```python
mu = 0.5
d = 0.125
sigma_n = math.sqrt(d / n)
```

```python
norm_5 = sts.norm(mu, sigma_n)
```

```python
plt.plot(x, norm_5.pdf(x), label='PDF(n=5)')
plt.hist(selected_avarage_arr_5, bins =80, range=(0,1), density=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```
