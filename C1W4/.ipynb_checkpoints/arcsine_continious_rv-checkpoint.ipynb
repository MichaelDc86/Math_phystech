### Arcsine continious random variable

импортируем необходимые python модули(я испоьлзую python3)

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
from functools import reduce
%matplotlib inline
```

[ссылка на wiki на распределение arcsine](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%B0%D1%80%D0%BA%D1%81%D0%B8%D0%BD%D1%83%D1%81%D0%B0)

[link](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.arcsine.html#scipy.stats.arcsine) ссылка на документуцию scipy.stats.arcsine


данные для выбранного распределения из wiki

$E[x]=0.5$

$D[x]=0.125$

создаем выборку объема 1000

```python
volume = 1000
sample = sts.arcsine.rvs(size=volume)
# len(sample)
# sample
```

histogram

```python
x = np.linspace(0,1,1000)  # массив точек для построения распредления

y = 1.0 / (math.pi * np.sqrt(x * (1 - x)))  # формула из wiki
plt.plot(x, y, '.', label='theoretical plot(from wiki)')  # теоретическая плотность по формуле из wiki

plt.plot(x, sts.arcsine.pdf(x), '+', label='theoretical pdf(arcsine.pdf)')  # теоретическая плотность по формуле из wiki

plt.hist(sample, bins=30, density=True, range=(0, 1), label='sample')  # строим нормированную гистограмму(в 3 питоне нормировка это density)
plt.ylabel('$f(x)$/number from sample')  # подписываем оси f(x)
plt.xlabel('$x$')
plt.legend(loc='best', frameon=False) # размещаем легенду 
plt.show()
```

Генерируем 1000 выборок объемов n    

### n = 5

```python
n = 5
selected_avarage_arr_5 = np.array([])  #массив для выборочных средних выборок объема 5
for i in range(1, 1001):  # генерируем 1000 выборок
    sample_ = sts.arcsine.rvs(size=n) 
    selected_avarage = sum(sample_) / len(sample_)  # среднее, как сумма, деленная на количество
    selected_avarage_arr_5 = np.append(selected_avarage_arr_5, selected_avarage)  # добавляем в массив
# len(selected_avarage_arr_5)
# selected_avarage_arr
```

параметры нормального распределения
$N(mu, sigma_n)$

```python
mu = 0.5  # из wiki
d = 0.125  # из wiki
sigma_n = math.sqrt(d / n)  # сигма для нормального распределния, приближающеего выборку средних
```

```python
norm_5 = sts.norm(mu, sigma_n)  # нромальное распределение, приближающеего выборку средних(согласно ЦПТ)
```

```python
plt.plot(x, norm_5.pdf(x), label='PDF(n=5)')  # плотность распределения
plt.hist(selected_avarage_arr_5, bins =80, range=(0,1), density=True)  # гистограмма средних для выборок объемом n
plt.ylabel('f(x)$/selected_avarage')
plt.xlabel('$x$')
plt.legend(loc='best', frameon=False)
```
