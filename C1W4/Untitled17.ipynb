```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Распределение Лапласа

Зададим параметры

```python
u = 2
b = 1
```

Сформируем выборку из распределения Лапласа (состоящую из 1000 элементов):

```python
laplace_rv = sts.laplace(u,b)
sample = laplace_rv.rvs(1000)
```

Сформируем гистограмму плотности распределения и поверх нее нарисуем график теоретической плотности распределения:

```python
plt.hist(sample,density=True) #Построение диаграммы
x = np.linspace(-10,10,1000)
pdf = laplace_rv.pdf(x)
plt.plot(x,pdf,label = 'Theoretical') #Построение графика
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

Построение гистограммы выборочного среднего для выборок размера N = 5

```python
from math import sqrt

N = 5

L = []
for i in range(1000):
    sample = laplace_rv.rvs(N)
    L.append(sum(sample)/N)

plt.hist(L,density=True) #Строим гистограмму 

#Коэффициенты норм. распределения
mu = u
sigma = 2/(b**2)
normal_rv = sts.norm(mu,sqrt(sigma/N))
x = np.linspace(-2.5,7.5,100)
pdf = normal_rv.pdf(x)
plt.plot(x,pdf)
plt.plot(x,pdf, label = 'Normal distribution')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

Построение гистограмм выборочного среднего для выборок размера N = 10,15

```python
N = 10

L = []
for i in range(1000):
    sample = laplace_rv.rvs(N)
    L.append(sum(sample)/N)

plt.hist(L,density=True) #Строим гистограмму 

#Коэффициенты норм. распределения
mu = u
sigma = 2/(b**2)
normal_rv = sts.norm(mu,sqrt(sigma/N))
x = np.linspace(-2.5,7.5,100)
pdf = normal_rv.pdf(x)
#Построение графика норм. распределения
plt.plot(x,pdf, label = 'Normal distribution')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```
